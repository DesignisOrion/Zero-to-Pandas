# Working with KDE Plots
KDE = Kernal Density Estimation

Allows us to estimate the probablity density function.
We will do this in a non-parametric way.

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg').dropna()

cars.shape
```

```python
cars.head()
```

# Univariate (1 variable) KDEplot

```python
sns.kdeplot(cars.horsepower);
```

```python
# Different arguements options for Seaborn
```

```python
# Shading
sns.kdeplot(cars.horsepower, shade=True);
```

```python
# Bandwidth
sns.kdeplot(cars.horsepower, bw=10);
```

```python
#cumulative 
sns.kdeplot(cars.horsepower, cumulative=True);
```

# Bivariate (2D) KDEplot

```python
# Working with horsepower and mpg (2D)
sns.kdeplot(cars.horsepower, cars.mpg);
```

```python
# You may want more rings or more different levels in the plog. use: n_levels
sns.kdeplot(cars.horsepower, cars.mpg, n_levels=20);
```

```python
# shaded version
sns.kdeplot(cars.horsepower, cars.mpg, shade=True);
```

```python
# if you want to turn off the shading on the lowest density, you can use the following:

# setting white grid style
sns.set_style('whitegrid')

sns.kdeplot(cars.horsepower, cars.mpg, shade=True, shade_lowest=False);
```

```python
# if you want to see what the density represents, you can use a color bar
sns.set_style('whitegrid')

sns.kdeplot(cars.horsepower, cars.mpg, shade=True, shade_lowest=False, cbar=True);
```
