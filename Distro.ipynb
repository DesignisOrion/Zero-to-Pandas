# Seaborn distplots

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg').dropna()
cars.shape
```

```python
cars.head()
```

```python
sns.set_style('white')
```

# Building the Plot

```python
#rugplot will be turned off by default
sns.distplot(cars.weight);
```

```python
# Want to turn off KDE to do just a histogram plot
# This is the only way you can do a histogram inside of seaborn.
sns.distplot(cars.weight, kde=False);
```

```python
# Turn off the histogram
sns.distplot(cars.weight, hist = False);
```

```python
# Turn on the rugplot
sns.distplot(cars.weight, hist=False, rug=True);
```

# Bins

```python
sns.distplot(cars.weight, bins = 7);
```

```python
# passing our own list to make unequal bins
sns.distplot(cars.weight, bins=[1500, 2000, 2500, 3000, 3500]);
```

# Fit

```python
from scipy.stats import norm #you can pass skewnorm or any other distribution
sns.distplot(cars.weight, fit = norm);
```

```python
# if you want to turn off kde
sns.distplot(cars.weight, fit = norm, kde=False);
```

# kwargs

# hist_kws

```python
# adjust the transparency
sns.distplot(cars.weight, kde=False, hist_kws={'alpha': 1});
```

```python
sns.distplot(cars.weight, kde=False, hist_kws={'alpha':1, 'histtype': 'step'});
```

```python
sns.distplot(cars.weight, kde=False, hist_kws={'alpha':1, 'histtype': 'stepfilled'});
```

# kde_kws

```python
sns.distplot(cars.weight, kde_kws={'lw': 5}); # keys are always string values
```

```python
sns.distplot(cars.weight, kde_kws={'bw': 100});
```

# rug_kws

```python
sns.distplot(cars.weight, rug=True);
```

```python
sns.distplot(cars.weight, rug=True, rug_kws={'height':0.15});
```

```python
# Chnage the color
sns.distplot(cars.weight, rug=True, color='r', rug_kws={'height':0.15});
```

```python
sns.distplot(cars.weight, rug=True, rug_kws={'height':0.15, 'color': 'xkcd:navy'});
```
