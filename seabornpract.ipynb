# Seaborn Practice Notes

```python
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Built in data sets for seaborn
print(sns.get_dataset_names())
```

```python
# loading car crash df
crash_df = sns.load_dataset('car_crashes')
crash_df
```

# Distribution plots

### Distribution Plot

```python
#The line is a kernel density estimation of the total result
sns.distplot(crash_df['not_distracted'])
```

```python
# Removes KDE
sns.distplot(crash_df['not_distracted'], kde=False)
```

```python
# bins are like buckets used to divide data between different intervals
sns.distplot(crash_df['not_distracted'], kde=False, bins=25)
```

```python
# Joint Plot
# Used to compare to distribution and plots a scatterplot by default
# In this we called for the x and the y axis. called the df we are using and placed a regression line.

sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

```python
# KDE jointplot
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='kde')
```

```python
# Hex distribution
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='hex')
```

### KDE plot

```python
sns.kdeplot(crash_df['alcohol'])
```

### Pair Plots

```python
# pair plots plots the relationships accross the entire df numerical value

sns.pairplot(crash_df)
```

```python
# Loading tips df for another example
tips_df = sns.load_dataset('tips')
tips_df
```

```python
# hue allows you to pass in categorical data and the charts will be colorized based on the data.
sns.pairplot(tips_df, hue='sex', palette='Blues')
```

### Rug Plot

```python
# Plot a single column of data points as sticks. The rug plot is not used much.
sns.rugplot(tips_df['tip'])
```

 # Categorical Plots

### Bar Plots

```python
# The bar is the variance
#sns.barplot(x='sex', y='total_bill', data=tips_df)

# Default it's aggrigated by the mean but you can change for :
# Standard deviation
sns.barplot(x='sex', y='total_bill', data=tips_df, estimator =np.std)
# Median
sns.barplot(x='sex', y='total_bill', data=tips_df, estimator =np.median)
# Variance
sns.barplot(x='sex', y='total_bill', data=tips_df, estimator =np.var)
# covariance
sns.barplot(x='sex', y='total_bill', data=tips_df, estimator =np.cov)
```

### Count Plot

```python
sns.countplot(x='sex', data=tips_df)
```

### Box Plot

```python
sns.boxplot(x='day', y = 'total_bill', data=tips_df, hue="sex")

# To move the legend
plt.legend(loc=0)
```

### Violin Plot
 A combo of box and kde plots

```python
sns.violinplot(x='day', y='total_bill', data=tips_df, hue='sex')
```

```python
# adding a split allows you to compare how the catagories compare to each other
sns.violinplot(x='day', y='total_bill', data=tips_df, hue='sex' ,split = True)
```

## Strip Plot


```python
plt.figure(figsize=(8,5))

sns.stripplot(x ='day', y='total_bill', data=tips_df)
```

```python
# jitter allows the plots to be scattered more instead of being on top of each other.

plt.figure(figsize=(8,5))
sns.stripplot(x ='day', y='total_bill', data=tips_df, jitter=True)
```

```python
# Show difference between men and women add hue
sns.stripplot(x ='day', y='total_bill', data=tips_df, jitter=True, hue='sex')
```

```python
# dodge separate the data of men and women completely
sns.stripplot(x ='day', y='total_bill', data=tips_df, jitter=True, hue='sex', dodge=True)
```

## Swarm Plot

```python
sns.swarmplot(x='day', y='total_bill', data=tips_df)
```
